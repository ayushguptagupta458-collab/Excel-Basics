1. Explain the differrence between the Relative, Absolute and Mixed cell referencing.

   - Relative cell referencing:-The cell reference changes automatically when you copy the formula to another cell.
   - Absolute cell referencing:-The row and column do not change, no matter where you copy the formula.
     Mixed cell referencing :- Partially fixed reference.
